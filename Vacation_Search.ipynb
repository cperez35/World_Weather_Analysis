```python
### II.4  Add the cites to a marker layer map with a pop-up marker for each city that includes:
##        Hotel Name, City, Country, Current Weather Description with Maximun Temparature.   
```

```python
## II.4a  Create the Hotel DF. 
hotel_df = preferred_cities_df[["City", "Country", "Max Temp", "Weather Description","Lat", "Lng"]].copy()
hotel_df["Hotel Name"] = ""
hotel_df.head()
```

```python
### II.4b  Configure search parameters 

 ## First Set the Parameters to be passed to the API Call
  #  Create a dictionary of parameters  
params = {"radius": 5000  # 5000 meters from location
          ,"type": "lodging" # get us a hotel
          ,"key": g_key # our api key
          } # Set the parameters to search for a hotel

### II.4c  Crawl through Google NearbySearch to get data based on our locations

for index, row in hotel_df.iterrows(): # for each row in the hotel_DF
    # Get the latitude and longitude.
    lat = row["Lat"]
    lng = row["Lng"]

    # Add the latitude and longitude to the params dictionary 
    # as values to the location key.
    params["location"] = f"{lat},{lng}"

    # Use the search term: "lodging" and our latitude and longitude.
    base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
    
    # Make request and get the JSON data from the request
    hotels = requests.get(base_url, params=params).json()
    try:
    # Grab the first hotel from the results and store the name.
        hotel_df.loc[index, "Hotel Name"] = hotels["results"][0]["name"]
    except(IndexError):
        print("Hotel not found... skipping.")
```

```python
## review
hotel_df.head()
#hotels  ## raw JSON
```

```python
### II.4d  Create the infobox template as required by the Google NearBySearch API for Marker Layers
  # combined two fields to show as one marker - current weather description and temp
  # renamed a field in the display - haha!
   
info_box_template = """
<dl>
<dt>Hotel Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
<dt>Current Weather</dt><dd>{Weather Description} & {Max Temp} °F</dd> 
</dl>
"""
```

```python
# inspect and review. 
info_box_template
```

```python
### II.4d1 
# Iterate through the DataFrame by row, and populate the info_box_template
hotel_info = [info_box_template.format(**row) for index, row in hotel_df.iterrows()]
```

```python
### ## II.4Final - Create the marker layer map

# Get X: the latitude and longitude.
locations = hotel_df[["Lat", "Lng"]]

# Create figure var with parameters: Add a center and Zoom.  30 deg north 31 east
fig= gmaps.figure(center=(30.0, 31.0), zoom_level=1.5)

# Create the marker layer with parameters set in the info_box_template. 
marker_layer = gmaps.marker_layer(locations, info_box_content=hotel_info)

# Add the marker layer
fig.add_layer(marker_layer)

# Call the figure to plot the data.
fig
```
